```python
def plotmatrix(figno, sep12, sep23, legal, title):
    plt.figure(figno, figsize=(12, 12))
    color= ['red' if l == 0 else 'green' for l in legal]
    plt.scatter(sep23, sep12, c=color, s=1000, marker='s')
    plt.xlim(35, 125)
    plt.ylim(35, 125)
    plt.title(title)
    plt.xlabel("Separation Sub1-Sub2")
    plt.ylabel("Separation Sub2-Sub3")
    plt.gca().invert_yaxis()
#    plt.set_tick_params(labeltop='on')
#    plt.tick_top()
    plt.savefig('plt' + str(figno) + '.png')
    plt.show()
    
```

```python
plotmatrix(1, stat_data['Separation S1-S2'], stat_data['Separation S2-S3'], stat_data['Dropout Legal'], "Separation Matrix with dropout specs")
```

Rounding and making separations format compatible with modeled data dataframe:

```python
def format_separation(x):
    y = x*2
    y = round(y)
    y = y*5
    return y
```

```python
separation_data["Separation S1-S2"] = separation_data["arr1_2"].apply(lambda x: format_separation(x))
separation_data["Separation S2-S3"] = separation_data["arr2_3"].apply(lambda x: format_separation(x))
```

```python
def plot_decision_boundaries(X, y, classifier):
    """Function to plot the decision boundaries of a classification model.
    This uses just the first two columns of the data for fitting 
    the model as we need to find the predicted value for every point in 
    scatter plot.
    
    One possible improvement could be to use all columns fot fitting
    and using the first 2 columns and median of all other columns
    for predicting.
    
    Adopted from:
    http://scikit-learn.org/stable/auto_examples/ensemble/plot_voting_decision_regions.html
    http://scikit-learn.org/stable/auto_examples/cluster/plot_kmeans_digits.html
    """
    #reduced_data = X[:, :2]
    #model = model_class(**model_params)
    #classifier.fit(reduced_data, y)

    #cmap_light = ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF'])
    #cmap_bold = ListedColormap(['#FF0000', '#00FF00', '#0000FF'])
    
    cmap_light = ListedColormap(['#AAFFAA', '#FFFFCC', '#FFAAAA'])
    cmap_bold = ListedColormap(['#00FF00', '#FFFF00', '#FF0000'])
    
    # Step size of the mesh. Decrease to increase the quality of the VQ.
    h = 0.1     # point in the mesh [x_min, m_max]x[y_min, y_max].    

    # Plot the decision boundary. For that, we will assign a color to each
    x_min, x_max = 35, 125
    y_min, y_max = 35, 125
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

    # Obtain labels for each point in mesh using the model.
    Z = classifier.predict(np.c_[xx.ravel(), yy.ravel()])    

    x_min, x_max = 35, 125
    y_min, y_max = 35, 125
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))

    Z = classifier.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)

    plt.figure(figsize=(12, 12))
    plt.gca().invert_yaxis()
    plt.contourf(xx, yy, Z, alpha=0.4, cmap=cmap_light)
    plt.scatter(X['Separation S1-S2'], X['Separation S2-S3'], c=y, alpha=0.8, cmap=cmap_bold)
    return plt
```

```python
plot_decision_boundaries(training_data, training_labels, classifier);
```
