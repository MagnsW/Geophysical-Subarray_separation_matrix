```python
import glob
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
statfiles = []
for file in glob.glob("./stats_files/*.txt"):
    statfiles.append(file)
```

```python
print(len(statfiles))
```

```python
print(statfiles)
```

```python
column_names = ['Farfield Name', 'Peak', 'Peak %ch', 'P-to-P', 'P-to-P %ch', 'P/B', 'P/B %ch', 'X-corr', 'AvgdB', 'MaxdB', 'MaxPhase', 'Coordinates', 'Dip/X', 'Az/Y', 'Dist/Z']
```

```python
stat_data = pd.DataFrame()
```

```python
skip = 15
```

```python
for stat in statfiles:
    temp_stat_data = pd.read_csv(stat, names=column_names, skiprows=skip, delim_whitespace=True)
    stat_data = stat_data.append(temp_stat_data)
    temp_stat_data = []
    
```

```python
stat_data
```

```python
stat_data['Separation S1-S2'] = stat_data.apply(lambda row: int(str(row['Farfield Name'])[16:19]), axis=1)
stat_data['Separation S2-S3'] = stat_data.apply(lambda row: int(str(row['Farfield Name'])[-3:]), axis=1)
```

```python
stat_data.dtypes
```

```python
stat_data 
```

```python
stat_data['Dropout Legal'] = (stat_data['Peak %ch'] <= 10) & (stat_data['P/B'] >= 28) & (stat_data['X-corr'] > 0.998)
```

```python
stat_data
```

```python
def plotmatrix(figno, sep12, sep23, legal, title):
    plt.figure(figno, figsize=(12, 12))
    color= ['red' if l == 0 else 'green' for l in legal]
    plt.scatter(sep23, sep12, c=color, s=1000, marker='s')
    plt.xlim(35, 125)
    plt.ylim(35, 125)
    plt.title(title)
    plt.xlabel("Separation Sub1-Sub2")
    plt.ylabel("Separation Sub2-Sub3")
    plt.gca().invert_yaxis()
#    plt.set_tick_params(labeltop='on')
#    plt.tick_top()
    plt.savefig('plt' + str(figno) + '.png')
    plt.show()
    
```

```python
plotmatrix(1, stat_data['Separation S1-S2'], stat_data['Separation S2-S3'], stat_data['Dropout Legal'], "Separation Matrix with dropout specs")
```

```python
condlist = [stat_data['X-corr'] >= 0.996, stat_data['X-corr'] >= 0.990, stat_data['X-corr'] < 0.990]
choicelist = ['green', 'yellow', 'red']
stat_data['X-corr Legal'] = (np.select(condlist, choicelist))
```

```python
stat_data
```

```python
def plotmatrix_tri(figno, sep12, sep23, legal, title):
    plt.figure(figno, figsize=(12, 12))
    color = legal
    plt.scatter(sep23, sep12, c=color, s=1000, marker='s')
    plt.xlim(35, 125)
    plt.ylim(35, 125)
    plt.title(title)
    plt.xlabel("Separation Sub2-Sub3")
    plt.ylabel("Separation Sub1-Sub2")
    plt.gca().invert_yaxis()
#    plt.set_tick_params(labeltop='on')
#    plt.tick_top()
    plt.savefig('plt' + str(figno) + '.png')
    plt.show()
```

```python
plotmatrix_tri(2, stat_data['Separation S1-S2'], stat_data['Separation S2-S3'], stat_data['X-corr Legal'], "Separation Matrix with x-corr specs")
```

```python
stat_data_condensed = stat_data[['Separation S1-S2', 'Separation S2-S3', 'X-corr']].reset_index()
```

```python
stat_data_pivot = stat_data_condensed.pivot('Separation S1-S2', 'Separation S2-S3', 'X-corr')
```

```python
stat_data_pivot
```

```python
def heatmap(figno, data):
    f, ax = plt.subplots(figsize=(18, 12))
    #pal = sns.color_palette(('red', 'yellow' ,'green'), n_colors=3)
    pal = sns.color_palette('RdYlGn', n_colors=3)
    sns.heatmap(data, vmin=0.984, vmax=1.002, annot=True, fmt='g', cmap=pal, linewidths=0.75, cbar_kws=dict(ticks=[0.984, .990, .996, 1]))
    plt.savefig('plt' + str(figno) + '.png')
    
```

```python
heatmap(3, stat_data_pivot)
```
