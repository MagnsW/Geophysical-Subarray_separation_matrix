## Applying Dropout specs to the statistics

Dropout specs applied:

<p>x-correlation: 0.998</p>
<p>Absolute P/B: 30</p>
<p>Max spectral change: +/- 3dB</p>

```python
stat_data['Dropout Legal'] = (stat_data['MaxdB'] <= 3) & (stat_data['P/B'] >= 30) & (stat_data['X-corr'] > 0.998)
```

```python
plt.figure(figsize=(12,8))
plt.axhline(0.998, color='red')
plt.axvline(3, color='red')
sns.regplot(x=stat_data["MaxdB"], y=abs(stat_data["X-corr"]), fit_reg=False)
sns.regplot(x=stat_data[(stat_data['Dropout Legal'])]["MaxdB"], y=stat_data[(stat_data['Dropout Legal'])]["X-corr"], fit_reg=False);
```

```python
plt.figure(figsize=(12,8))
plt.axhline(0.998, color='red')
plt.axvline(30, color='red')
sns.regplot(x=stat_data["P/B"], y=abs(stat_data["X-corr"]), fit_reg=False)
sns.regplot(x=stat_data[(stat_data['Dropout Legal'])]["P/B"], y=stat_data[(stat_data['Dropout Legal'])]["X-corr"], fit_reg=False);
```

```python
def plotmatrix(figno, sep12, sep23, legal, title):
    plt.figure(figno, figsize=(12, 12))
    color= ['red' if l == 0 else 'green' for l in legal]
    plt.scatter(sep23, sep12, c=color, s=1000, marker='s')
    plt.xlim(35, 125)
    plt.ylim(35, 125)
    plt.title(title)
    plt.xlabel("Separation Sub1-Sub2")
    plt.ylabel("Separation Sub2-Sub3")
    plt.gca().invert_yaxis()
#    plt.set_tick_params(labeltop='on')
#    plt.tick_top()
    plt.savefig('plt' + str(figno) + '.png')
    plt.show()
    
```

```python
plotmatrix(1, stat_data['Separation S1-S2'], stat_data['Separation S2-S3'], stat_data['Dropout Legal'], "Separation Matrix with dropout specs")
```

Below, the x-corr spec is defining green (>=0.996), yellow (>=0.990) or red (0.990) combinations in of subarray separations

```python
condlist = [stat_data['X-corr'] >= 0.996, stat_data['X-corr'] >= 0.990, stat_data['X-corr'] < 0.990]
choicelist = ['green', 'yellow', 'red']
stat_data['X-corr Legal'] = (np.select(condlist, choicelist))
```

## Automatic QC Based on String Separation TXT File

Reading in subarray separation txt file to dataframe

```python
separation_file = "./Online_separation/seq0104_substring_separation.txt"
```

```python
separation_data = pd.DataFrame()
```

```python
separation_data = pd.read_csv(separation_file, skiprows=1, delim_whitespace=True)
```

```python
separation_data
```

Plotting Subarray separation per shot:

```python
sns.set_style("darkgrid")
plt.figure(figsize=(17,12))
sns.regplot(x=separation_data["shotpno"], y=separation_data["arr1_2"], fit_reg=False, label="arr1-2")
sns.regplot(x=separation_data["shotpno"], y=separation_data["arr2_3"], fit_reg=False, label="arr2-3")
plt.legend();
```

Rounding and making separations format compatible with modeled data dataframe:

```python
def format_separation(x):
    y = x*2
    y = round(y)
    y = y*5
    return y
```

```python
#Testing function
print(format_separation(5.75))
```

```python
separation_data["Separation S1-S2"] = separation_data["arr1_2"].apply(lambda x: format_separation(x))
separation_data["Separation S2-S3"] = separation_data["arr2_3"].apply(lambda x: format_separation(x))
```

```python
separation_data
```

Merging separation data with modeled data:

```python
separation_data_merge = separation_data.merge(stat_data_clean, how='inner', left_on=['Separation S1-S2', 'Separation S2-S3'], right_on=['Separation S1-S2', 'Separation S2-S3']).sort_values(by=['shotpno']).reset_index().drop(['index', 'Farfield Name'], axis=1)
```

```python
separation_data_merge
```

Plotting shots and their legal label. There are two separate legal labels, one for x-correlation (green, yellow, red) and one based on dropout (binary)

```python
def plot_shots(shotno, sep12, sep23, legal, title):
    plt.figure(figsize=(17, 12))
    sns.set_style("darkgrid")
    color = legal
    plt.scatter(shotno, sep12, c=color)
    plt.scatter(shotno, sep23, c=color)
    plt.xlabel("Shot point number")
    plt.ylabel("Subarray separation (dm)")
    plt.title(title)
    plt.show()
```

```python
plot_shots(separation_data_merge["shotpno"], separation_data_merge["Separation S1-S2"], separation_data_merge["Separation S2-S3"], separation_data_merge["X-corr Legal"], "X-correlation criteria" )
```

```python
plot_shots(separation_data_merge["shotpno"], separation_data_merge["Separation S1-S2"], separation_data_merge["Separation S2-S3"], separation_data_merge["Dropout Legal"], "Dropout criteria" )
```
